```python
with open('9.txt', 'r') as f:
    s = f.read().split("\n")
# s='''R 4
# U 4
# L 3
# D 1
# R 4
# D 1
# L 5
# R 2
# '''.split("\n")
s.pop()
num = 2
rope = [[0,0] for _ in range(num)]

spots = set([(0,0)])

def apart(head, tail):
    return abs(head[0]-tail[0]) >= 2 or abs(head[1]-tail[1]) >= 2
    
for line in s:
    direction, steps = line.split(" ")
    
    for _ in range(int(steps)):
        if direction=="R":
            rope[-1][0] += 1
            for i in reversed(range(1, num)):
                head, tail = rope[i], rope[i-1]
                if apart(head, tail):
                    tail[0] = head[0]-1
                    tail[1] = head[1]
                
        if direction=="U":
            rope[-1][1] += 1
            for i in reversed(range(1, num)):
                head, tail = rope[i], rope[i-1]
                if apart(head, tail):
                    tail[1] = head[1]-1
                    tail[0] = head[0]
                
        if direction=="L":
            rope[-1][0] -= 1
            for i in reversed(range(1, num)):
                head, tail = rope[i], rope[i-1]
                if apart(head, tail):
                    tail[0] = head[0]+1
                    tail[1] = head[1]
                
        if direction=="D":
            rope[-1][1] -= 1
            for i in reversed(range(1, num)):
                head, tail = rope[i], rope[i-1]
                if apart(head, tail):
                    tail[1] = head[1]+1
                    tail[0] = head[0]
#         print(rope)
        spots.add(tuple(rope[0]))
                
#         print(head, tail)
print(len(spots))
```
