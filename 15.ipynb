```python
from collections import Counter

with open("15.txt", "r") as f:
    s = f.read().split("\n")
s.pop()

sensors, beacons = [], []

for line in s:
    elems = line.split("=")
    sensors.append((int(elems[1][:-3]), int(elems[2].split(":")[0])))
    beacons.append((int(elems[3][:-3]), int(elems[4])))
    
intervals = []
y = 2000000

for s,b in zip(sensors, beacons):
    man_dist = abs(s[0]-b[0])+abs(s[1]-b[1])
    man_dist = man_dist-abs(s[1]-y)
    if man_dist <= 0:
        continue
    intervals.append((s[0]-man_dist, s[0]+man_dist))
    
intervals.sort(key=lambda x:x[0])

gaps = []
last = intervals[0][1]
min_point, max_point = float("inf"), float("-inf")

for start, end in intervals:
    min_point, max_point = min(min_point, start), max(max_point, end)
    if start > last + 1:
        gaps.append((last+1, start-1))
    last = max(last, end)
    
res = max_point-min_point+1
for start, end in gaps:
    res -= end-start
for beacon in set(beacons):
    if beacon[1] == y and min_point <= beacon[0] <= max_point:
        res -= 1
print(res)
```

```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
import numpy as np

class Line:
    def __init__(self, x1,y1,m):
        self.start = (x1,y1)
        self.m = m
        
    def intersect(self, line2):
        start = line2.start
        x1,y1,m1 = self.start[0], self.start[1],self.m
        x2,y2,m2 = line2.start[0], line2.start[1], line2.m
        
        if m1 == m2:
            return -1,-1
        
        x = round((m1*x1-y1-m2*x2+y2)/(m1-m2))
        y = m1*(x-x1)+y1
        
        if not(0<=x<=4000000 and 0<=y<=4000000):
            return -1,-1
        
        return x,y

with open("15.txt", "r") as f:
    s = f.read().split("\n")
s.pop()

sensors, beacons = [], []

for line in s:
    elems = line.split("=")
    sensors.append((int(elems[1][:-3]), int(elems[2].split(":")[0])))
    beacons.append((int(elems[3][:-3]), int(elems[4])))
    
lines = []

for s,b in zip(sensors, beacons):
    man_dist = abs(s[0]-b[0]) + abs(s[1]-b[1])
    x1,y1 = s[0]-man_dist, s[1]
    x2,y2 = s[0], s[1]+man_dist
    x3,y3 = s[0]+man_dist, s[1]
    x4,y4 = s[0], s[1]-man_dist
    
    line1 = Line(x1,y1,-1)
    line2 = Line(x2,y2,1)
    line3 = Line(x3,y3,-1)
    line4 = Line(x4,y4,1)
    x = np.array(list(range(21)))   
    
    lines += [line1, line2, line3, line4]

points = set()
for i in range(len(lines)):
    for j in range(i+1, len(lines)):
        point = lines[j].intersect(lines[i])
        if point[0]!=-1:
            points.add(point)
                
            
poss = set()

for point in points:
    x,y = point
    if (x+1,y+1) in points and (x+1, y-1) in points and (x+2, y) in points:
        poss.add((x+1,y))
for point in poss:
    x,y = point
    valid = True
    for s,b in zip(sensors, beacons):
        man_dist = abs(s[0]-b[0]) + abs(s[1]-b[1])
        man_dist2 = abs(s[0]-x) + abs(s[1]-y)
        if man_dist2<=man_dist:
            valid=False
            break
    if valid:
        print(x,y, x*4000000+y)
```
